```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
for i in df1.columns:
  df1[i] = encoder.fit_transform(df1[i])
```

```python
from sklearn.model_selection import train_test_split
x = df1.drop('y',axis=1)
y = df1['y']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
df1.info()
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state = 60)
rf.fit(x_train,y_train)
```

```python
features = rf.feature_importances_
features
```

```python
important_feature = x.columns[features>0.06]
important_feature
```

```python
#smote
from imblearn.over_sampling import SMOTE
smt = SMOTE(random_state = 42)
x_train_smt,y_train_smt = smt.fit_resample(x_train,y_train)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x1_train = scaler.fit_transform(x_train_smt)
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.preprocessing import StandardScaler

L_R = LogisticRegression()
L_R.fit(x1_train,y_train_smt)
pred_LR = L_R.predict(x_test)

accuracy = accuracy_score(y_test,pred_LR)
conf_matrix = confusion_matrix(y_test, pred_LR)
class_report = classification_report(y_test, pred_LR)
print('accuracy:', accuracy)
print('confusion matrix:\n', conf_matrix)
print('classification report:\n', class_report)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=100)
clf.fit(x1_train,y_train_smt)
y_pred_DC = clf.predict(x_test)
accuracy = accuracy_score(y_test,y_pred_DC)
conf_matrix = confusion_matrix(y_test,y_pred_DC)
class_report = classification_report(y_test,y_pred_DC)
print("Accuracy:",accuracy)
print("Confusion_matrix:\n",conf_matrix)
print("Classification_report:\n",class_report)
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(random_state = 100)
random_forest.fit(x1_train,y_train_smt)
y_pred_rf = random_forest.predict(x_test)
accuracy_rf = accuracy_score(y_test,y_pred_rf)
conf_matrix = confusion_matrix(y_test,y_pred_rf )
class_report = classification_report(y_test, y_pred_rf)
print('accuracy:', accuracy)
print('confusion matrix:\n', conf_matrix)
print('classification report:\n', class_report)
```
